```python
import pandas as pd
from sklearn.model_selection import train_test_split

df = pd.read_csv("data/station_day_processed.csv")
```

```python
# Drop rows where regression target is NaN
df = df.dropna(subset=['AQI'])

# Features and targets
X = df.drop(columns=['AQI','AQI_Bucket','Date'])
y_reg = df['AQI']          # regression target
y_clf = df['AQI_Bucket']   # classification target
```

```python
# Split data
X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(X, y_reg, test_size=0.2, random_state=42)
X_train_clf, X_test_clf, y_train_clf, y_test_clf = train_test_split(X, y_clf, test_size=0.2, random_state=42)
```

```python
from sklearn.impute import SimpleImputer

X_train_reg = X_train_reg.apply(pd.to_numeric, errors='coerce')
X_test_reg = X_test_reg.apply(pd.to_numeric, errors='coerce')
X_train_clf = X_train_clf.apply(pd.to_numeric, errors='coerce')
X_test_clf = X_test_clf.apply(pd.to_numeric, errors='coerce')
```

```python
# Impute missing features
num_imputer = SimpleImputer(strategy='median')
X_train_reg = num_imputer.fit_transform(X_train_reg)
X_test_reg = num_imputer.transform(X_test_reg)
X_train_clf = num_imputer.fit_transform(X_train_clf)
X_test_clf = num_imputer.transform(X_test_clf)
```

```python
#Scale features
from sklearn.preprocessing import StandardScaler

scaler_reg = StandardScaler()
X_train_reg_scaled = scaler_reg.fit_transform(X_train_reg)
X_test_reg_scaled = scaler_reg.transform(X_test_reg)

scaler_clf = StandardScaler()
X_train_clf_scaled = scaler_clf.fit_transform(X_train_clf)
X_test_clf_scaled = scaler_clf.transform(X_test_clf)
```

```python
# Regression Model
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

reg_model = LinearRegression()
reg_model.fit(X_train_reg_scaled, y_train_reg)
```

```python

y_train_pred = reg_model.predict(X_train_reg_scaled)
y_test_pred = reg_model.predict(X_test_reg_scaled)
```

```python

print("Regression Metrics (AQI):")
print("Train R²:", r2_score(y_train_reg, y_train_pred))
print("Test R²:", r2_score(y_test_reg, y_test_pred))
print("Test MAE:", mean_absolute_error(y_test_reg, y_test_pred))
# print("Test RMSE:", mean_squared_error(y_test_reg, y_test_pred, squared=False))
```

```python
# Classification Model
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
# Drop rows where classification target is NaN
clf_df = df.dropna(subset=['AQI_Bucket'])
X_clf = clf_df.drop(columns=['AQI','AQI_Bucket','Date'])
y_clf = clf_df['AQI_Bucket']
```

```python
X_train_clf, X_test_clf, y_train_clf, y_test_clf = train_test_split(
    X_clf, y_clf, test_size=0.2, random_state=42
)
```

```python
# Encode target
le = LabelEncoder()
y_train_clf_enc = le.fit_transform(y_train_clf)
y_test_clf_enc = le.transform(y_test_clf)
```

```python
# Impute & scale already done above
clf_model = RandomForestClassifier(n_estimators=100, random_state=42)
clf_model.fit(X_train_clf_scaled, y_train_clf_enc)

y_test_clf_pred = clf_model.predict(X_test_clf_scaled)
```

```python

print("\nClassification Metrics (AQI_Bucket):")
print("Accuracy:", accuracy_score(y_test_clf_enc, y_test_clf_pred))
print(classification_report(y_test_clf_enc, y_test_clf_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test_clf_enc, y_test_clf_pred))
```
